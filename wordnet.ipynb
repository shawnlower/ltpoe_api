```python
import requests
import rdflib
from rdflib import Graph, plugin, Namespace
from rdflib.store import Store
import pandas as pd
from io import StringIO
```

```python
resp = requests.get('http://wordnet-rdf.princeton.edu/ttl/lemma/mathematics')
```

```python
g = Graph()
```

```python
g.load(StringIO(resp.text), format='ttl')
```

```python
set(g.predicates())
```

```python
ontology = Graph()
ontology.load(StringIO(requests.get('http://wordnet-rdf.princeton.edu/ontology').text))
len(ontology)
```

```python
set([ e.toPython() for e in ontology.objects()])
```
